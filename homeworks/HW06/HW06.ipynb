```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, f1_score, roc_auc_score,
                             classification_report, confusion_matrix,
                             roc_curve, precision_recall_curve, average_precision_score)
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (RandomForestClassifier, AdaBoostClassifier,
                              GradientBoostingClassifier, StackingClassifier)
from sklearn.inspection import permutation_importance
import joblib
import json
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('S06-hw-dataset-04.csv')
print("Размер датасета:", df.shape)
print("\nПервые 5 строк:")
display(df.head())

print("Информация о датасете:")
df.info()

print("\nСтатистики:")
display(df.describe())

print("\nПропуски:")
print(df.isnull().sum())

target_dist = df['target'].value_counts(normalize=True)
print("Распределение целевой переменной:")
print(f"Класс 0: {target_dist[0]:.2%} ({df['target'].value_counts()[0]})")
print(f"Класс 1: {target_dist[1]:.2%} ({df['target'].value_counts()[1]})")

plt.figure(figsize=(8, 6))
df['target'].value_counts().plot(kind='bar')
plt.title('Распределение классов (target)')
plt.xlabel('Класс')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.savefig('artifacts/figures/target_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

X = df.drop(['target', 'id'], axis=1)  # исключаем target и id
y = df['target']

print(f"Признаки: {X.shape[1]}")
print(f"Примеры признаков: {list(X.columns[:5])}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y,
    shuffle=True
)

print(f"Train размер: {X_train.shape}")
print(f"Test размер: {X_test.shape}")
print(f"\nРаспределение в train: {np.bincount(y_train) / len(y_train)}")
print(f"Распределение в test: {np.bincount(y_test) / len(y_test)}")

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name=""):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

    metrics = {
        'model': model_name,
        'accuracy': accuracy_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred, average='binary'),
    }

    if y_pred_proba is not None:
        metrics['roc_auc'] = roc_auc_score(y_test, y_pred_proba)
        metrics['avg_precision'] = average_precision_score(y_test, y_pred_proba)

    # Дополнительные метрики
    print(f"\n{model_name}:")
    print(classification_report(y_test, y_pred))

    return metrics

dummy = DummyClassifier(strategy='most_frequent', random_state=42)
metrics_dummy = evaluate_model(dummy, X_train, X_test, y_train, y_test, "Dummy (most_frequent)")

logreg_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'))
])
metrics_logreg = evaluate_model(logreg_pipeline, X_train, X_test, y_train, y_test, "LogisticRegression")

print("Decision Tree Classifier")
dt = DecisionTreeClassifier(random_state=42)

# Параметры для GridSearch
param_grid_dt = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_leaf': [1, 2, 5, 10],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# Используем StratifiedKFold для дисбаланса
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_dt = GridSearchCV(
    dt, param_grid_dt,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
grid_dt.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_dt.best_params_}")
print(f"Лучший ROC-AUC (CV): {grid_dt.best_score_:.4f}")

# Оценка на тесте
dt_best = grid_dt.best_estimator_
metrics_dt = evaluate_model(dt_best, X_train, X_test, y_train, y_test, "DecisionTree (optimized)")

print("Random Forest Classifier")

rf = RandomForestClassifier(random_state=42, n_jobs=-1, class_weight='balanced_subsample')

# Упрощенная сетка параметров для скорости
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', 0.5]
}

grid_rf = GridSearchCV(
    rf, param_grid_rf,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
grid_rf.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_rf.best_params_}")
print(f"Лучший ROC-AUC (CV): {grid_rf.best_score_:.4f}")

rf_best = grid_rf.best_estimator_
metrics_rf = evaluate_model(rf_best, X_train, X_test, y_train, y_test, "RandomForest (optimized)")


print("Gradient Boosting Classifier")


gb = GradientBoostingClassifier(random_state=42)

param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0]
}

grid_gb = GridSearchCV(
    gb, param_grid_gb,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
grid_gb.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_gb.best_params_}")
print(f"Лучший ROC-AUC (CV): {grid_gb.best_score_:.4f}")

gb_best = grid_gb.best_estimator_
metrics_gb = evaluate_model(gb_best, X_train, X_test, y_train, y_test, "GradientBoosting (optimized)")


print("Stacking Classifier")

# Базовые модели для стекинга
base_models = [
    ('dt', DecisionTreeClassifier(max_depth=5, min_samples_leaf=10, random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)),
    ('gb', GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42))
]

# Мета-модель
meta_model = LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')

stacking = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=cv,
    n_jobs=-1
)

stacking.fit(X_train, y_train)
metrics_stacking = evaluate_model(stacking, X_train, X_test, y_train, y_test, "StackingClassifier")


all_metrics = [metrics_dummy, metrics_logreg, metrics_dt, metrics_rf, metrics_gb, metrics_stacking]
metrics_df = pd.DataFrame(all_metrics)

print("Сравнение всех моделей:")
display(metrics_df.sort_values('roc_auc', ascending=False))

plt.figure(figsize=(10, 8))
models = {
    'LogisticRegression': logreg_pipeline,
    'DecisionTree': dt_best,
    'RandomForest': rf_best,
    'GradientBoosting': gb_best,
    'Stacking': stacking
}

for name, model in models.items():
    if hasattr(model, 'predict_proba'):
        y_pred_proba = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        auc = roc_auc_score(y_test, y_pred_proba)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random (AUC = 0.5)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривые для всех моделей')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.savefig('artifacts/figures/roc_curves.png', dpi=300, bbox_inches='tight')
plt.show()

best_model_name = metrics_df.loc[metrics_df['roc_auc'].idxmax(), 'model']
best_model = models.get(best_model_name, rf_best)  # по умолчанию RF

y_pred_best = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class 0', 'Class 1'],
            yticklabels=['Class 0', 'Class 1'])
plt.ylabel('Фактический класс')
plt.xlabel('Предсказанный класс')
plt.title(f'Confusion Matrix: {best_model_name}')
plt.savefig('artifacts/figures/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(10, 8))

for name, model in models.items():
    if hasattr(model, 'predict_proba'):
        y_pred_proba = model.predict_proba(X_test)[:, 1]
        precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
        avg_precision = average_precision_score(y_test, y_pred_proba)
        plt.plot(recall, precision, label=f'{name} (AP = {avg_precision:.3f})')

# Baseline для дисбаланса
baseline = y_test.mean()
plt.plot([0, 1], [baseline, baseline], 'k--', label=f'Baseline (AP = {baseline:.3f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall кривые')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.savefig('artifacts/figures/pr_curves.png', dpi=300, bbox_inches='tight')
plt.show()


print(f"Permutation Importance для {best_model_name}")

result = permutation_importance(
    best_model, X_test, y_test,
    n_repeats=10,
    random_state=42,
    scoring='roc_auc',
    n_jobs=-1
)

sorted_idx = result.importances_mean.argsort()[::-1]

top_n = min(15, len(X.columns))
plt.figure(figsize=(12, 8))
plt.barh(range(top_n), result.importances_mean[sorted_idx][:top_n][::-1])
plt.yticks(range(top_n), np.array(X.columns)[sorted_idx][:top_n][::-1])
plt.xlabel('Уменьшение ROC-AUC')
plt.title(f'Permutation Importance (топ-{top_n})')
plt.tight_layout()
plt.savefig('artifacts/figures/feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nТоп-{top_n} наиболее важных признаков:")
for i in range(top_n):
    idx = sorted_idx[i]
    print(f"{i+1:2d}. {X.columns[idx]:30s}: {result.importances_mean[idx]:.4f} ± {result.importances_std[idx]:.4f}")

# 1. Сохранение метрик
metrics_df.to_csv('artifacts/metrics_test.csv', index=False)
metrics_df.to_json('artifacts/metrics_test.json', orient='records', indent=2)

# 2. Сохранение лучших параметров
search_results = {
    'DecisionTree': {
        'best_params': grid_dt.best_params_,
        'best_score': grid_dt.best_score_
    },
    'RandomForest': {
        'best_params': grid_rf.best_params_,
        'best_score': grid_rf.best_score_
    },
    'GradientBoosting': {
        'best_params': grid_gb.best_params_,
        'best_score': grid_gb.best_score_
    }
}

with open('artifacts/search_summaries.json', 'w') as f:
    json.dump(search_results, f, indent=2)

# 3. Сохранение лучшей модели
joblib.dump(best_model, 'artifacts/best_model.joblib')

# 4. Метаданные лучшей модели
best_model_meta = {
    'model_name': best_model_name,
    'model_type': str(type(best_model)),
    'best_params': search_results.get(best_model_name, {}).get('best_params', {}),
    'test_metrics': metrics_df[metrics_df['model'] == best_model_name].iloc[0].to_dict(),
    'feature_importance_top5': {
        X.columns[sorted_idx[i]]: float(result.importances_mean[sorted_idx[i]])
        for i in range(min(5, len(sorted_idx)))
    }
}

with open('artifacts/best_model_meta.json', 'w') as f:
    json.dump(best_model_meta, f, indent=2)
```
